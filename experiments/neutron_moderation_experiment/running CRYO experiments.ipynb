```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from moderation_experiment_DT_CRYO import Moderation_Experiment_DTCRYO
from moderation_experiment_DD_CRYO import Moderation_Experiment_DDCRYO

from neutrons_code_L import Capsule
from neutrons_code_L.functions import neutron_counting, widths
```

```python
import scienceplots
plt.style.use(['science', 'ieee', 'no-latex'])
plt.rcParams.update({'figure.dpi': '300'})
```

```python
experiment_DT = Moderation_Experiment_DTCRYO(n_runs=30, time=320)  # t[320] = 80ps so just after peak burn
experiment_DD = Moderation_Experiment_DDCRYO(n_runs=30, time=320)  # t[320] = 80ps so just after peak burn

experiment_DT.run_experiment()
experiment_DD.run_experiment()
```

```python
# 0 moderation case

En = np.geomspace(1e-5, 16, 500)[::-1]

unmod_params_DT = np.array([[1e30, 0.5, 0.5, 0, 0, 35e-6],
                            [1e32, 0.5, 0.5, 0, 0, 1e-5],
                            [1e32, 0, 0, 0, 1, 1e-5]])


unmod_params_DD = np.array([[1e30, 0, 1, 0, 0, 35e-6],
                            [1e32, 0, 1, 0, 0, 1e-5],
                            [1e32, 0, 0, 0, 1, 1e-5]])

no_targets = [None for _ in range(3)]

unmod_DT = Capsule(En, unmod_params_DT, no_targets)
unmod_DT.solve_neutron_transport()
results_unmod_DT = neutron_counting(unmod_DT, experiment_DT.time)


unmod_DD = Capsule(En, unmod_params_DD, no_targets)
unmod_DD.solve_neutron_transport()
results_unmod_DD = neutron_counting(unmod_DD, experiment_DD.time)
```

```python
experiment = experiment_DT

fig,axs = plt.subplots(4,4, figsize=(20,15))

x_vals = experiment.H_fraction_hotspot

for i,x in enumerate(['1MeV', '100keV', '10keV', '1keV']):

    # axs[0,i] = hotspot
    axs[0,i].plot(x_vals, experiment.final_results_hotspot[f's{x}_0'])
    axs[0,i].plot(x_vals, experiment.final_results_shell[f's{x}_0'])
    axs[0,i].plot(x_vals, experiment.final_results_ablator[f's{x}_0'])


    # axs[1,i] = shell
    axs[1,i].plot(x_vals, experiment.final_results_hotspot[f's{x}_1'])
    axs[1,i].plot(x_vals, experiment.final_results_shell[f's{x}_1'])
    axs[1,i].plot(x_vals, experiment.final_results_ablator[f's{x}_1'])

    # axs[2,i] = ablator
    axs[2,i].plot(x_vals, experiment.final_results_hotspot[f's{x}_2'])
    axs[2,i].plot(x_vals, experiment.final_results_shell[f's{x}_2'])
    axs[2,i].plot(x_vals, experiment.final_results_ablator[f's{x}_2'])


    axs[3,i].plot(x_vals, experiment.final_results_hotspot[f's{x}_3'])
    axs[3,i].plot(x_vals, experiment.final_results_shell[f's{x}_3'])
    axs[3,i].plot(x_vals, experiment.final_results_ablator[f's{x}_3'])



    axs[0,i].plot([0], results_unmod_DT[f's{x}_0'], 'x', color='k')
    axs[1,i].plot([0], results_unmod_DT[f's{x}_1'], 'x', color='k')
    axs[2,i].plot([0], results_unmod_DT[f's{x}_2'], 'x', color='k')
    axs[3,i].plot([0], results_unmod_DT[f's{x}_3'], 'x', color='k')



axs[0,0].set_ylabel('Hotspot')
axs[1,0].set_ylabel('Shell')
axs[2,0].set_ylabel('Ablator')
axs[3,0].set_ylabel('Outside')

axs[0,0].set_title('less than 1MeV')
axs[0,1].set_title('less than 100keV')
axs[0,2].set_title('less than 10keV')
axs[0,3].set_title('less than 1keV')
```

```python
# what is the best fusion fuel to try and moderate?

fig,ax = plt.subplots(figsize=(5,5))

energy_range = 's100keV'

y = experiment_DT.final_results_shell[f'{energy_range}_1'] / np.float64(results_unmod_DT[f'{energy_range}_1'])
ax.plot(experiment_DT.N_hydrogen/1e17, y, '-', label='DT fuel')

y = experiment_DD.final_results_shell[f'{energy_range}_1'] / np.float64(results_unmod_DD[f'{energy_range}_1'])
ax.plot(experiment_DD.N_hydrogen/1e17, y, '-', label='DD fuel')

ax.legend()

ax.set_ylabel('increase in flux')
ax.set_xlabel(r'$10^{17}$ hydrogen atoms')


# ax.plot([0], results_unmod_DT[f'{energy_range}_1'], 'x', color='k')
# ax.plot([0], results_unmod_DD[f'{energy_range}_1'], 'x', color='k')

plt.ylabel('increase of neutron flux below 100keV')
plt.xlabel(r'$10^{17}$ hydrogen atoms')

# plt.savefig('/Users/brynlloyd/Desktop/CRYO_appendix_plot.png', dpi=600)
```
